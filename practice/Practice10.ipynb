### 7. Train a decision tree for wine dataset and test it
- load wine dataset in scikit learn (hint:load_wine)
- split the dataset into training and test data (7:3 ratio), set random_state to 0
- train a decision tree with training dataset
- visualize the trained tree
- test the trained tree on test dataset
- store test labels to the variable 'y_pred2'

```python
# TO DO
y_pred2=np.zeros_like(y_test)

# 와인 데이터셋 로드
wine = sklearn.datasets.load_wine()

# train, test 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3, random_state=0)

# Decision Tree 모델 훈련
decision_tree_classifier_wine = DecisionTreeClassifier()
decision_tree_classifier_wine.fit(X_train, y_train)

# 훈련된 Decision Tree 모델 시각화
plt.figure(figsize=(12, 12))
plot_tree(decision_tree_classifier_wine, feature_names=wine.feature_names, class_names=wine.target_names, filled=True)
plt.show()

# 테스트 데이터셋으로 예측
y_pred2 = decision_tree_classifier_wine.predict(X_test)
```

### Print accuracy (do not modify the following block)

```python
print('Accuracy: %.2f' % sklearn.metrics.accuracy_score(y_test, y_pred2))
```

### 8. Load wine dataset from scikit learn, train the logistic regression, and change at least one hyperparameter

```python
wine = sklearn.datasets.load_wine()


X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3, random_state=0)

from sklearn.linear_model import LogisticRegression
logistic_regression_model = LogisticRegression(C=0.01)
logistic_regression_model.fit(X_train, y_train)

y_pred_logistic_regression = logistic_regression_model.predict(X_test)
```

```python
print('Accuracy: %.2f' % sklearn.metrics.accuracy_score(y_test, y_pred_logistic_regression))
```

### 9. Load iris dataset from scikit learn and train the logistic regression, and change at least one hyperparameter

```python
iris = sklearn.datasets.load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=0)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

logistic_regression_model = LogisticRegression(C=10)
logistic_regression_model.fit(X_train_scaled, y_train)

y_pred_logistic_regression = logistic_regression_model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_logistic_regression)
print("Accuracy of the modified model:", accuracy)
```

### 10. Load breast cancer dataset from scikit learn, train the logistic regression, and change at least one hyperparameter

```python
cancer = sklearn.datasets.load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=0)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

logistic_regression_model = LogisticRegression(C=0.01)
logistic_regression_model.fit(X_train_scaled, y_train)

y_pred_logistic_regression = logistic_regression_model.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred_logistic_regression)
print("Accuracy of the modified model:", accuracy)
```
