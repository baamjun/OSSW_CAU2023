```python
import numpy as np

import matplotlib.pyplot as plt 
%matplotlib inline
```

### 1. importing required libraries and loading the iris plants dataset (classification)
- libraries: datasets, model_selection, metrics, tree, ensemble

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 2. dividing the datasets into two parts i.e. training datasets and test datasets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### 3. creating a random forest classifier 

```python
random_forest_model = RandomForestClassifier(n_estimators=10)
```

### 4. training the model on the training dataset 

```python
random_forest_model.fit(X_train, y_train)
```

### 5. performing predictions on the test dataset and measuring the performance
- metrics are used to find accuracy or error

```python
y_pred_random_forest = random_forest_model.predict(X_test)
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)
print(f"Random Forest Accuracy: {accuracy_random_forest}")
```

### 6. visualizing all trees
- RandomForestClassifier.estimators_
- tree.plot_tree

```python
from sklearn.tree import plot_tree

estimators = random_forest_model.estimators_

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 18))

for i, ax in enumerate(axes.ravel()):
    if i < len(estimators):
        plot_tree(estimators[i], feature_names=iris.feature_names, class_names=iris.target_names, filled=True, ax=ax)
        ax.set_title(f'Tree {i+1}')
    else:
        ax.axis('off')

plt.show()
```

### 7. creating make moons dataset, splitting them, and training the model on the training dataset
- datasets.make_moons

```python
from sklearn.datasets import make_moons

# make moons dataset
X, y = make_moons(n_samples=1000, noise=0.2)

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# training
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# evaluation
y_pred_decision_tree = decision_tree_model.predict(X_test)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
print(f"Decision Tree Accuracy: {accuracy_decision_tree}")
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

model = RandomForestClassifier(n_estimators=4)
model.fit(X_train, y_train)

fig, axes = plt.subplots(nrows = 1,ncols = 5,figsize = (20,4))
for index in range(0, 4):
    DecisionBoundaryDisplay.from_estimator(
            model.estimators_[index], X, alpha=0.5, response_method="predict", ax=axes[index]
        )
    axes[index].scatter(X[:,0],X[:,1],c=y)
    axes[index].set_title('Tree: ' + str(index))

DecisionBoundaryDisplay.from_estimator(
        model, X, alpha=0.5, response_method="predict", ax=axes[-1]
    )
axes[-1].scatter(X[:,0],X[:,1],c=y)
axes[-1].set_title('Forest: ' + str(index))    
plt.show()
```

### 9. 
1. load wine dataset from scikit learn, 
2. train both single decision tree and a random forest
3. compare their accuracy

```python
# load data
wine = datasets.load_wine()
X = wine.data
y = wine.target

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Decision Tree 모델 생성 및 학습
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# Random Forest 모델 생성 및 학습
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train, y_train)

# 정확도 비교
y_pred_decision_tree = decision_tree_model.predict(X_test)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
print(f"Decision Tree Accuracy: {accuracy_decision_tree}")

y_pred_random_forest = random_forest_model.predict(X_test)
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)
print(f"Random Forest Accuracy: {accuracy_random_forest}")
```

### 10. 
1. load Olivetti faces dataset from scikit learn

```python
# load data
olivetti = datasets.fetch_olivetti_faces(shuffle=True)
X = olivetti.data
y = olivetti.target

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Decision Tree 모델 생성 및 학습
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# Random Forest 모델 생성 및 학습
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train, y_train)

# 정확도 비교
y_pred_decision_tree = decision_tree_model.predict(X_test)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
print(f"Decision Tree Accuracy: {accuracy_decision_tree}")

y_pred_random_forest = random_forest_model.predict(X_test)
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)
print(f"Random Forest Accuracy: {accuracy_random_forest}")
```

### 11. 
1. load Olivetti faces dataset from scikit learn
    - sklearn.datasets.fetch_olivetti_faces
2. train two random forests with different hyperparameters
3. visualize them
4. compare their accuracy

```python
# load data
olivetti = datasets.fetch_olivetti_faces(shuffle=True)
X = olivetti.data
y = olivetti.target

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 두 개의 Random Forest 모델 생성 및 학습
    # 첫 번째 랜덤 포레스트
random_forest_model_1 = RandomForestClassifier(n_estimators=50, max_depth=10) # 하이퍼파라미터 다르게
random_forest_model_1.fit(X_train, y_train)

    # 두 번째 랜덤 포레스트
random_forest_model_2 = RandomForestClassifier(n_estimators=100, max_depth=20) # 하이퍼파라미터 다르게
random_forest_model_2.fit(X_train, y_train)


# 두 개의 랜덤 포레스트 시각화
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

    # 첫 번째 랜덤 포레스트
axes[0].set_title('Random Forest 1')
plot_tree(random_forest_model_1.estimators_[0], filled=True, rounded=True, ax=axes[0])

    # 두 번째 랜덤 포레스트
axes[1].set_title('Random Forest 2')
plot_tree(random_forest_model_2.estimators_[0], filled=True, rounded=True, ax=axes[1])

plt.show()

# 정확도 비교
y_pred_random_forest_1 = random_forest_model_1.predict(X_test)
accuracy_random_forest_1 = accuracy_score(y_test, y_pred_random_forest_1)
print(f"Random Forest 1 Accuracy: {accuracy_random_forest_1}")

y_pred_random_forest_2 = random_forest_model_2.predict(X_test)
accuracy_random_forest_2 = accuracy_score(y_test, y_pred_random_forest_2)
print(f"Random Forest 2 Accuracy: {accuracy_random_forest_2}")
```

```python
# 첫 번째 랜덤 포레스트 이미지 및 예측 결과 시각화
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(15, 6))

for i, ax in enumerate(axes.ravel()):
    ax.imshow(X_test[i].reshape(64, 64), cmap='gray')
    ax.set_title(f"True: {y_test[i]}\nPredicted: {y_pred_random_forest_1[i]}", fontsize=10)
    ax.axis('off')

plt.show()

print(f"Random Forest 1 Accuracy: {accuracy_random_forest_1}")
```

```python
# 두 번째 랜덤 포레스트 이미지 및 예측 결과 시각화
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(15, 6))

for i, ax in enumerate(axes.ravel()):
    ax.imshow(X_test[i].reshape(64, 64), cmap='gray')
    ax.set_title(f"True: {y_test[i]}\nPredicted: {y_pred_random_forest_2[i]}", fontsize=10)
    ax.axis('off')

plt.show()

print(f"Random Forest 2 Accuracy: {accuracy_random_forest_2}")
```
