### Problem 7.
Write a Python programming to display a bar chart of the popularity of programming Languages.

Sample data:
- Programming languages: Java, Python, PHP, JavaScript, C#, C++
- Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

__The code must give the output shown in the following screenshot:__


```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]

# TODO
# plot bar graph
plt.bar(x_pos, popularity, color='blue')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")
plt.xticks(x_pos, x)

# TODO
# Turn on the grid
plt.grid(True, linestyle='--', alpha=0.6, color='r')
# Customize the minor grid
plt.minorticks_on()
plt.grid(which='minor', color='gray', linestyle='-', linewidth=0.5, alpha=0.2)

plt.show()
```

### Problem 8.
Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

Sample Data:
- Means (men) = (22, 30, 35, 35, 26)
- Means (women) = (25, 32, 30, 35, 29)

__The code must give the output shown in the following screenshot:__


```python
# data to plot
n_groups = 5
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

# TODO
# Plot bar graph for men's data 
men_bars = plt.bar(index + bar_width/2, men_means, bar_width, alpha=opacity, color='g', label="Men")

# Plot bar graph for women's data
women_bars = plt.bar(index + bar_width/2 + bar_width, women_means, bar_width, alpha=opacity, color='r', label="Women")

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()
plt.tight_layout()
plt.show()
```

### Problem 9.
Write a Python program to create bar plots with error bars on the same figure.
__The code must give the output shown in the following screenshot:__


```python
N = 5
menMeans = (54.74, 42.35, 67.37, 58.24, 30.25)
menStd = (4, 3, 4, 1, 5)
# the x locations for the groups
ind = np.arange(N)    
# the width of the bars
width = 0.35   

#TODO
# plot bar graph with error bar
plt.bar(ind+0.38, menMeans, width, yerr=menStd,color='r', ecolor='b', label='Men')

plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity')

plt.xticks(np.arange(0, N) + 0.19, ['0.0', '0.5', '1.0', '1.5', '2.0'])

# TODO
# Turn on the grid
plt.grid(True, linestyle='-', alpha=0.6, color = 'g')
# Customize the minor grid
plt.minorticks_on()
plt.grid(which='minor', color='gray', linestyle='--', linewidth=0.5, alpha=0.2)

plt.show()
```
